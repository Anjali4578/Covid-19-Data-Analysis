## Importing another dataset

```python
df2.tail()
```

### Comparing death and recovered cases in countries with the least death cases

```python
# Sort dataset which contains only top 15 countries

sorted_df2 = df2.sort_values(by = "Deaths")
sorted_df2 = sorted_df2[:15]
```

```python
# Plotting the comparison

plt.figure(figsize = (10, 6))
plt.bar(sorted_df2["Country/Region"], sorted_df2["Confirmed"], label = 'Confirmed', color = 'green')
plt.bar(sorted_df2["Country/Region"], sorted_df2["Active"], label = 'Active', color = 'red')

plt.legend()
plt.xlabel('Country')
plt.ylabel('The number of cases')
plt.xlim()
plt.ylim()
plt.title('Comparison of Confirmed and Active cases')
plt.xticks(rotation = 45)
plt.show()
```

### Comparing recovered and death cases

```python
recovered = df2["Recovered"].sum()
death = df2["Deaths"].sum()

data = [recovered, death]
labels = ["REcovered cases", "Deaths"]

plt.pie(data, labels = labels, autopct = "%.2f%%")
```

### Comparing new recovered and new death cases

```python
new_recovered = df2["Recovered"].sum()
new_death = df2["Deaths"].sum()

data = [new_recovered, new_death]
labels = ["New Recovered cases", "New Deaths"]

plt.pie(data, labels = labels, autopct = "%.2f%%")
```

## Importing the worldometer dataset

```python
df3 = pd.read_csv("worldometer_data.csv")
```

```python
df3.head()
```

#### Creating a dataframe to identify ratio of Total deaths to total population

```python
ratio = round(df3['TotalDeaths'] / df3["TotalCases"] * 100, 2)
data = {
    "Country" : df3["Country/Region"],
    "Death precent" : ratio
}

pd.DataFrame(data)
```

#### Creating pie chart to observe serious/critical cases

```python
critical = df3["Serious,Critical"].sum(), (df3["ActiveCases"] - df3["Serious,Critical"]).sum()
labels = ["Serious/Critical cases", "Not serious cases"]

plt.pie(critical, labels = labels, autopct = "%2.f%%", shadow = True)
```

### Continents with the number of Active cases in the form of pie chart 

```python
continental_active = df3[["Continent", "ActiveCases"]]
continental_active = df3.groupby("Continent").sum()

continental_active
```

```python
plt.pie(continental_active["ActiveCases"], labels = continental_active.index, autopct = "%.2f%%", 
        pctdistance = 0.85, explode = [0, 0, 0.35, 0, 0, 0], shadow = True, counterclock = False)
```

### Comparing total death and total recovered

```python
plt.figure(figsize = (10, 6))
plt.bar(df3["Country/Region"][:15], df3["TotalRecovered"][:15], label = 'Total Recovered', color = 'green')
plt.bar(df3["Country/Region"][:15], df3["TotalDeaths"][:15], label = 'Total Deaths', color = 'red')

plt.legend()
plt.xlabel('Country')
plt.ylabel('The number of cases')
plt.title('Comparison of Total Recovered and Total Deaths')
plt.xticks(rotation = 45)
plt.show()
```

### Importing another dataset

```python
df4 = pd.read_csv("country_wise_latest.csv")
```

```python
df4.head()
```

### Death cases by countries

```python
import pandas as pd
import plotly.express as px

data = {
    "Country/Region" : df4["Country/Region"],
    "NewDeaths" : df4["New deaths"]
}
```

```python
df = pd.DataFrame(data)

# Read the shapefile for world countries
world = px.data.gapminder()
```

```python
# Merge the shapefile with the data
merged = world.merge(df, left_on = 'country', right_on = 'Country/Region', how = 'left')
```

```python
fig = px.choropleth(merged, locations = 'iso_alpha', color = 'NewDeaths', color_continuous_scale = "RdPu", 
                    range_color = (0, df['NewDeaths'].max()), labels = {'NewDeaths': 'New Deaths'}, hover_name = 'country')

fig.update_geos(showcountries = True, countrycolor = "darkgrey", showcoastlines = True, coastlinecolor = "lightgrey", 
               showland = True, landcolor = "white", showocean = True, oceancolor = "lightblue", showlakes = True, lakecolor = "lightblue")

fig.update_layout(title_text = 'New Deaths by Country', title_x = 0.5)

fig.show()
```

### Active cases by countries

```python
import pandas as pd
import plotly.express as px

data = {
    "Country/Region" : df4["Country/Region"],
    "Active" : df4["Active"]
}

df = pd.DataFrame(data)
world = px.data.gapminder()
```

```python
merged = world.merge(df, left_on = 'country', right_on = 'Country/Region', how = 'left')

fig = px.choropleth(merged, locations = 'iso_alpha', color = 'Active', color_continuous_scale = "RdPu", 
                    range_color = (0, df['Active'].max()), labels = {'Active': 'Active'}, hover_name = 'country')

fig.update_geos(showcountries = True, countrycolor = "darkgrey", showcoastlines = True, coastlinecolor = "lightgrey", 
               showland = True, landcolor = "white", showocean = True, oceancolor = "lightblue", showlakes = True, lakecolor = "lightblue")

fig.update_layout(title_text = 'Active cases by Country', title_x = 0.5)

fig.show()
```

###  One week change by countries

```python
import pandas as pd
import plotly.express as px

data = {
    "Country/Region" : df4["Country/Region"],
    "1 week change" : df4["1 week change"]
}

df = pd.DataFrame(data)
world = px.data.gapminder()

merged = world.merge(df, left_on = 'country', right_on = 'Country/Region', how = 'left')

fig = px.choropleth(merged, locations = 'iso_alpha', color = '1 week change', color_continuous_scale = "RdPu", 
                    range_color = (0, df['1 week change'].max()), labels = {'1 week change': '1 week change'}, hover_name = 'country')

fig.update_geos(showcountries = True, countrycolor = "darkgrey", showcoastlines = True, coastlinecolor = "lightgrey", 
               showland = True, landcolor = "white", showocean = True, oceancolor = "lightblue", showlakes = True, lakecolor = "lightblue")

fig.update_layout(title_text = 'One week change by Country', title_x = 0.5)

fig.show()
```
